```python
g = sns.FacetGrid(omit_data, col="Facility Type")
g.map(plt.hist, "Total Project Cost");
```

```python
omit_data["Project Status Date"] = (pd.to_datetime(omit_data["Project Status Date"])).dt.year
g2 = sns.FacetGrid(omit_data, col="Facility Type", hue="Project Status Date")
g2.map(plt.scatter, "Total Project Cost", "Project Status Date", alpha=.7)
g2.add_legend();
```
