```python
sentence = input('Sentence: ')
subject = input('subject: ')
verb = input('verb: ')
adjective = input('adjective: ')
object = input('object: ')

newsubject = input('new subject: ')
newverb = input('new verb: ')
newadjective = input('new adjective: ')
newobject = input('new object: ')

if subject!='':
    sentence =sentence.replace(subject, newsubject)
if verb!='':
    sentence =sentence.replace(verb, newverb)
if adjective!='':
    sentence =sentence.replace(adjective, newadjective)
if object!='':
    sentence =sentence.replace(object, newobject)
    
print('new sentecne is '+ sentence)
```

```python
import random

targetNum = random.randint(0,100)
print('User generate the random number is',targetNum)

def guessFunction():
    guessNum = 50;
    
    while True:
        print('Program guess number is',guessNum)
        userAnswer = input('The guess number is("too high" or "too low" or "correct") : ')
        if userAnswer == 'too high':
            result = False
            guessNum -=1
        elif  userAnswer == 'too low':
            result = False
            guessNum +=1
        elif userAnswer == 'correct':
            result = True
            print('Correct guess number is',guessNum)
        if(result):
            break
            
guessFunction()
    
```
